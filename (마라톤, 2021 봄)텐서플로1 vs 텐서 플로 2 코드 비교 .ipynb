```python
# Simple Linear Regression using Tensorflow2
```

```python
import numpy as np
import tensorflow as tf
```

```python
# X and Y data
x_train = [1,2,3,4]  # 1 dimension,  input_dim = 1
y_train = [1,2,3,4]

model = tf.keras.models.Sequential()
# keras  layer   .      

model.add(tf.keras.layers.Dense(1,input_dim=1))
#   layer .
# Dense(output, input)

# optimizer
sgd = tf.keras.optimizers.SGD(lr=0.01) # lr learning rate
# optimizer    SGD. 
# SGD Stochastic gradient descent optimizer .

# cost/loss function
model.compile(loss='mean_squared_error',optimizer=sgd)
# mean_squared_error  mse  .

# Fit the line
model.fit(x_train,y_train,epochs=2000) # 2000 

# Testing our model
print(model.predict(np.array([5])))

```
