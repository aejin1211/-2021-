```python
import tensorflow.compat.v1 as tf  #version1을 tf로 지정하겠습니다.
tf.disable_v2_behavior()  #version2는 사용하지 않겠습니다.
#tf.session()함수는 이전 텐서플로우 1에서 사용되는 방식이여서 텐서플로우 1의 함수를 쓰기 위해
```

```python
a = tf.Variable(2)
b = tf.Variable(3)
c = tf.add(a, b)
print(c)
```

```python
sess = tf.Session()
```

```python
sess.run(c)
```

```python
# 실행이 되지 않음, 왜냐 변수를 선언하고 나서 initializing을 해줘야 하기 때문  
```

```python
a = tf.Variable(2)
b = tf.Variable(3)
c = tf.add(a,b)
initial = tf.global_variables_initializer()
print(c)
```

```python
sess = tf.Session()
```

```python
sess.run(initial)
sess.run(c)
```
